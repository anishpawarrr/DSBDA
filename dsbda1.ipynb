```python
import pandas as pd
```

```python
ds = pd.read_csv("pseudo_facebook.csv")
ds.head()
```

```python
sub1 = ds.iloc[:,[0,1,5]]
sub1
```

```python
sub2 = ds.iloc[:,[0,9]]
sub2
```

```python
sub1 = sub1.sort_values(by = "userid",ascending= True)
sub2 = sub2.sort_values(by = "userid",ascending= True)
```

```python
sub2 = sub2[sub2["likes"] > 50]
sub2
```

```python
sub1
```

```python
mer = pd.merge(sub1, sub2, how='inner', on="userid")
mer
```

```python
transpose = mer.transpose()
transpose
```

```python
meltt = pd.melt(mer, id_vars=["userid"], value_vars=["likes", "age"])
meltt
```
